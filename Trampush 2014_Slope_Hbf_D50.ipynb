### An empirical law for bankfull characteristics relating Slope, bankfull depth and cahnnel-fill grain size.

####    logS= α0 + α1 log D50 + α2 log Hbf
          where α0= -2.08, α1= 0.254, α2= -1.09


Reference:
Trampush, S. M., Huzurbazar, S. & McElroy, B. Empirical assessment of theory for bankfull characteristics of alluvial channels. Water Resour Res. 50, 9211–9220 (2014).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Trampush law empirical parameters
a0 = -2.08
a1 = 0.2540
a2 = -1.09

def Slope(D50,Hbf):
    S = np.power(10,a0+a1*np.log10(D50)+a2*np.log10(Hbf))
    return S
```

```python
#Slope versus clast diameter (constang bankfull depth)
for Hbf_range in range(2,8,1):
    D50_range = np.linspace(0.01, 0.05,100)
    S = Slope(D50_range, Hbf_range)
    # plot the function
    plt.plot(D50_range,S, label = Hbf_range)

#plot study case High Amalgamated data
D50 = [0.022, 0.018, 0.019, 0.0185]
Hbf = [4.6, 4.2, 4.5, 5.8]
S_HA = Slope(D50, Hbf)
plt.scatter(D50, S_HA, label = 'High Amalgamated')

#plot study case Low Amalgamated data
D50 = [0.018, 0.018, 0.016]
Hbf = [3.2, 3.5, 3.9]
S_LA = Slope(D50, Hbf)
plt.scatter(D50, S_LA, label = 'Low Amalgamated')


#plot lines
plt.title("Trampush et al. (2014) empirical law")
plt.xlabel("D50(m)")
plt.ylabel("Slope(m/m)")
plt.legend(title='Bankfull depth',  loc = 'best', bbox_to_anchor=(1, 1))
plt.annotate("+ HBf", xy=(0.03, 0.0014), xytext=(0.03, 0.0007), ha="center", arrowprops=dict(arrowstyle="<-"))
plt.savefig('Trampush_S_D50.png', bbox_inches='tight', dpi=100)

# show the plot
plt.show()
```

```python
#Slope versus bankfull depth (constang grain size)
for n in range(10):
    D50_range = 0.01+n*0.005
    Hbf_range = np.linspace(2, 8,100)
    S = Slope(D50_range,Hbf_range)
    # plot the function
    plt.plot(Hbf_range,S, label = round(D50_range,3))

#plot study case Escanilla Fm High Amalgamated data
D50 = [0.022, 0.018, 0.019, 0.0185]
Hbf = [4.6, 4.2, 4.5, 5.8]
S_HA = Slope(D50,Hbf)
plt.scatter(Hbf, S_HA, label = 'High amalgamated')

#plot study case Escanilla Fm Low Amalgamated data
D50 = [0.018, 0.018, 0.016]
Hbf = [3.2, 3.5, 3.9]
S_LA = Slope(D50,Hbf)
plt.scatter(Hbf, S_LA, label = 'low amalgamated')

plt.xlabel("Bankfull depth(m)")
plt.ylabel("Slope(m/m)")
plt.legend(title='Grain size D50', loc = 'best', bbox_to_anchor=(1, 1))
plt.title("Trampush et al. (2014) empirical law")
plt.annotate("+ grain size", xy=(2.2, 0.001), xytext=(3.1, 0.0014), ha="center", arrowprops=dict(arrowstyle="<-"))
plt.savefig('Trampush_S_Hbf.png', bbox_inches='tight', dpi=100)
# show the plot
plt.show()
```
